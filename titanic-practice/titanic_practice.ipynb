```python
import pandas as pd
```

```python
data=pd.read_excel('titanic3.xls')
```

How many passengers are in the dataset, and how many columns?

```python
#there is 1309 passenger and 14 columns
data.shape
```

What percentage of passengers survived overall?

```python
#0.3819709702062643 % of passengers survived
data.survived.mean()
```

What was the survival rate for men vs women? (use groupby)

```python
#0.727468 for females and 0.190985 for males
data.groupby('sex').survived.mean()
```

What was the average age of passengers, and how many rows have a missing age?

```python
#average age is 30
data.age.mean()
```

```python
#263 missing values
data.age.isnull().sum()
```

Fill the missing ages with the median age (not mean — think about why median might be better, write your reasoning in your notes)

```python
#median is better because for mean we can have more older people than younger and the mean will be big but median takes the middle of all ages of passangers 
data["age"]=data.age.fillna(data.age.median())
```

Which passenger class (1st, 2nd, 3rd) had the highest survival rate?

```python
#first class had the highest survival rate
data.groupby('pclass').survived.mean()
```

Create a new column called family_size by adding SibSp + Parch + 1

```python
data["family_size"]=data.sibsp+data.parch+1
```

Did larger families survive more or less than people traveling alone? (group by family_size or a simple alone/not-alone flag)

```python
#we have a curve going from 1 to 4 family members the survival rate increases but passing 4 it startes to decrease
data.groupby('family_size').survived.agg(['mean','count'])
```
